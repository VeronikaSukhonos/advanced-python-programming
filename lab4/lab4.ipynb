# Завдання

Створити власний декоратор для кешування результатів функції – якщо певна функція вже запускалась с такими параметрами – не запускати функцію, а відобразити результати із кешу. Протестувати на 3-4 функціях з різною кількістю та типом параметрів.

```python
def cache_results(original_func):
  cache = {}

  def wrapper(*args, **kwargs):
    key = (*args, *kwargs.items())

    if key in cache:
      print('Getting results from cache...')
      return cache[key]

    print('Calling function...')
    cache[key] = original_func(*args, **kwargs)
    return cache[key]

  return wrapper
```

## Тестування

```python
@cache_results
def calc_sum(a, b=0):
  return a + b

print(calc_sum(5, 10))
print(calc_sum(5, 10))
print(calc_sum(5, 10))
print(calc_sum(30, -5))
print(calc_sum(30, -5))
print(calc_sum(1, 0))
print(calc_sum(1))
```

```python
@cache_results
def get_product_list(*args):
  return sorted([*args])

list1 = get_product_list('potato', 'juice', 'banana')
print(list1)
list2 = get_product_list('potato', 'juice', 'banana')
print(list2)
list3 = get_product_list('potato', 'juice')
print(list3)
print()
print(
  f'list1 is list2: {list1 is list2}',
  f'list1 is list3: {list1 is list3}',
  f'list2 is list3: {list2 is list3}',
  sep='\n'
)
```

```python
@cache_results
def get_person(name, age):
  return {'name': name, 'age': age, 'adult': age >= 18}

print(get_person('Bob', 15))
print(get_person('Alice', 20))
print(get_person('Bob', 15))
print(get_person('Alice', 20))
```

```python
@cache_results
def format_msg(greet='Hello', name='World', punct='!'):
  return f'{greet}, {name}{punct}'

print(format_msg(greet='Hey', name='Alice'))
print(format_msg(greet='Hey', name='Alice'))
print(format_msg(greet='Hey', name='Alice', punct=')'))
print(format_msg(name='Bob'))
print(format_msg(name='Bob'))
```

### Визначення допоміжних функцій для порівняння часу виконання скриптів та форматування виводу списків

```python
NP, IT = 'np', 'it'
time = {}

def compare_time():
  n, i = time[NP], time[IT]

  res = 'Conclusion: NumPy functions were '
  if (np.allclose(n, i)):
    res += 'AS effective as '
  elif (n < i):
    res += 'MORE effective than '
  elif (n > i):
    res += 'LESS effective than '
  res += f'pure Python iteractive constructions (diff = {abs(n - i)} s)'

  print(f'\nExecution time: NumPy = {n} s, pure Python = {i} s')
  print(res)

def print_l(l):
  N = len(l)

  if (N > 0 and type(l[0]) == list):
    for i in range(N):
      print(f"{'[' if i == 0 else ' '}{l[i]}{']' if i == N - 1 else ''}")
  else:
    print(l)
```

### Вправа 2

Вводиться число `n`. Розставити 1 та 0 у шаховому порядку, починаючи з 1 в матриці розміром `n` на `n`, використовуючи слайсинг.

```python
def task2_np(n):
  s = timer()

  matrix = np.ones((n, n))
  matrix[1::2, 0::2] = 0
  matrix[0::2, 1::2] = 0

  time[NP] = timer() - s
  print(matrix)


def task2_it(n):
  s = timer()

  matrix = [[1 for _ in range(n)] for _ in range(n)]
  for row in range(n):
    matrix[row][(row % 2 == 0)::2] = [0] * len(matrix[row][(row % 2 == 0)::2])

  time[IT] = timer() - s
  print_l(matrix)


task2_np(5)
print()
task2_it(5)
compare_time()
```

### Вправа 20

Згенерувати вектор із `n` випадкових елементів, що лежать в інтервалі (0,1). Замінити максимальний елемент на 0.

```python
def task20_np(n):
  s = timer()

  init_vector = np.random.random(n)
  res_vector = np.where(init_vector == np.max(init_vector), 0, init_vector)

  time[NP] = timer() - s
  print(f'Initial vector: {init_vector}')
  print(f'Changed vector: {res_vector}')


def task20_it(n):
  s = timer()

  init_vector = [random.random() for _ in range(n)]
  max_num = init_vector[0]
  for num in init_vector:
    if num > max_num:
      max_num = num
  res_vector = [0 if num == max_num else num for num in init_vector]

  time[IT] = timer() - s
  print(f'Initial vector: {init_vector}')
  print(f'Changed vector: {res_vector}')


task20_np(5)
print()
task20_it(5)
compare_time()
```

Ввести числа `n` і `m`. Згенерувати матрицю `n` на `n` випадкових цілих елементів з діапазону `[0, m]`. Замінити всі максимальні елементи цієї матриці на протилежні за знаком.

```python
def task21_np(n, m):
  s = timer()

  init_matrix = np.random.randint(0, m + 1, (n, n))
  max_num = np.max(init_matrix)
  res_matrix = np.where(init_matrix == max_num, -init_matrix, init_matrix)

  time[NP] = timer() - s
  print(f'Initial matrix:\n {init_matrix}')
  print(f'Changed matrix:\n {res_matrix}')


def task21_it(n, m):
  s = timer()

  init_matrix = [[random.randint(0, m) for _ in range(n)] for _ in range(n)]
  max_num = init_matrix[0][0]
  for row in init_matrix:
    for num in row:
      if num > max_num:
        max_num = num
  res_matrix = [[-num if num != max_num else num for num in row] for row in init_matrix]

  time[IT] = timer() - s
  print(f'Initial matrix:')
  print_l(init_matrix)
  print(f'Changed matrix:')
  print_l(res_matrix)


task21_np(4, 10)
print()
task21_it(4, 10)
compare_time()
```

## Завдання 2

Розв’язати систему лінійних алгебраїчних рівнянь за допомогою формул Крамера і виконати перевірку за допомогою:
  * матричного множення
  * оберненої матриці
  * функції `numpy.linalg.solve()`

Порівняти всі рішення за допомогою функції `numpy.allclose()`.

**Дана система рівнянь**:

$$
\begin{cases}
x_{1} + x_{2} - x_{3} - x_{4} = 0 \\
x_{2} + 2x_{3} - x_{4} = 2 \\
x_{1} - x_{2} - x_{4} = -1 \\
- x_{1} + 3x_{2} - 2x_{3} = 0
\end{cases}
$$

### Розв'язання СЛАР за формулами Крамера

```python
A = np.array([
  [1, 1, -1, -1],
  [0, 1, 2, -1],
  [1, -1, 0, -1],
  [-1, 3, -2, 0]
])

B = np.array([0, 2, -1, 0])

def cramer(A, B):
  det = np.linalg.det(A)

  if (det == 0):
    raise ValueError("Matrix of coefficients A has a zero determinant. Cannot use the Cramer's rule.")

  N, X = len(A), []

  for col in range(N):
    tmp = [[A[i][j] if j != col else B[i] for j in range(N)] for i in range(N)]
    tmp_det = np.linalg.det(tmp)
    X.append(round(tmp_det / det, 10))
  return X

X = cramer(A, B)
print("Solution vector X obtained using the Cramer's rule:", X)
```

### Перевірка знайденого розв’язку

```python
B_mult = A.dot(X)
print('Vector of free terms B_mult obtained by mutiplying matrix of coefficients A by X:', B_mult)
print('Vector of free terms B equals B_mult:', np.allclose(B, B_mult))

print()

X_inv = np.linalg.inv(A).dot(B)
print("Solution vector X_inv obtained by multiplying inversed matrix A⁻¹ by B:", X)
print("Solution vector X equals X_inv:", np.allclose(X, X_inv))

print()

X_np = np.linalg.solve(A, B)
print("Solution vector X_np obtained by using a dedicated NumPy function:", X_np)
print("Solution vector X equals X_np:", np.allclose(X, X_np))
```

## Завдання 3

Обчислити значення матричного виразу 2 способами:
  * з використанням універсальних функцій бібліотеки NumPy
  * за допомогою ітеративних конструкцій (з використанням циклів, спискових включень тощо)

Порівняти всі рішення за допомогою функції `numpy.allclose()`.

Для обох випадків підрахувати час виконання скрипту та зробити висновок.

**Даний матричний вираз**:

$$
A(A^{2} - B) - 2(B + A)B\text{, де }
A =
\begin{pmatrix}
2 & 3 & 1 \\
-1 & 2 & 4 \\
5 & 3 & 0
\end{pmatrix}
B =
\begin{pmatrix}
2 & 7 & 13 \\
-1 & 0 & 5 \\
5 & 13 & 21
\end{pmatrix}
$$

```python
A = [
  [2, 3, 1],
  [-1, 2, 4],
  [5, 3, 0]
]

B = [
  [2, 7, 13],
  [-1, 0, 5],
  [5, 13, 21]
]

def solve_matrix_expression_np(A, B):
  s = timer()

  res = np.subtract(np.dot(A, np.subtract(np.dot(A, A), B)), np.dot(np.multiply(np.add(B, A), 2), B))
  # res = A @ (A @ A - B) - 2 * (B + A) @ B

  time[NP] = timer() - s
  print(f'Result of the matrix expression using NumPy functions:\n{res}')
  return res

def mult_m(m1, m2):
  m1_rows, m1_cols, m2_rows, m2_cols = len(m1), len(m1[0]), len(m2), len(m2[0])

  if m1_cols != m2_rows:
    raise ValueError('Number of columns in the 1st matrix must equal number of rows in the 2nd matrix for multiplication.')

  res = [[0] * m2_cols for _ in range(m1_rows)]

  for i in range(m1_rows):
    for j in range(m2_cols):
      for k in range(m2_rows):
        res[i][j] += m1[i][k] * m2[k][j]
  return res

def add_m(m1, m2, add=True):
  m1_rows, m1_cols, m2_rows, m2_cols = len(m1), len(m1[0]), len(m2), len(m2[0])

  if not (m1_rows == m1_cols == m2_rows == m2_cols):
    raise ValueError('Matrices must have equal dimensions for addition.')

  res = [[0] * m1_rows for _ in range(m1_rows)]

  for i in range(m1_rows):
    for j in range(m2_cols):
      res[i][j] = m1[i][j] + m2[i][j] if add else m1[i][j] - m2[i][j]
  return res

def mult_m_by_num(m, num):
  m_rows, m_cols = len(m), len(m[0])

  res = [[0] * m_rows for _ in range(m_rows)]

  for i in range(m_rows):
    for j in range(m_cols):
      res[i][j] = m[i][j] * num
  return res

def solve_matrix_expression_it(A, B):
  s = timer()

  res = add_m(mult_m(A, add_m(mult_m(A, A), B, False)), mult_m(mult_m_by_num(add_m(B, A), 2), B), False)

  time[IT] = timer() - s
  print('Result of the matrix expression using pure Python iterative constructions:')
  print_l(res)
  return res

res_np = solve_matrix_expression_np(np.array(A), np.array(B))
print()
res_it = solve_matrix_expression_it(A, B)

print('\nResults are equal:', np.allclose(res_np, res_it))
compare_time()
```
